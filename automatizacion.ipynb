```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from openpyxl import Workbook
from openpyxl.drawing.image import Image

# 1Ô∏è‚É£ üìä Crear un DataFrame de ejemplo
data = pd.read_excel("datos_limpios.xlsx")

df = pd.DataFrame(data)

# 2Ô∏è‚É£ üîÑ Crear una tabla pivote (Ventas por Fecha y Categor√≠a)
pivot_df = df.pivot_table(values='Ventas', index='Fecha', columns='Categor√≠a', aggfunc='sum', fill_value=0)

# 3Ô∏è‚É£ üìà Generar y guardar gr√°ficos

# Gr√°fico de l√≠neas
plt.figure(figsize=(8, 4))
sns.lineplot(data=pivot_df, marker='o')
plt.title('Ventas por Categor√≠a a lo Largo del Tiempo')
plt.xlabel('Fecha')
plt.ylabel('Ventas')
plt.xticks(rotation=45)
plt.legend(title='Categor√≠a')
plt.tight_layout()
plt.savefig("grafico_lineas.png")  # Guardar imagen

# Gr√°fico de barras
pivot_df.plot(kind='bar', figsize=(8, 4))
plt.title('Ventas por Categor√≠a')
plt.ylabel('Ventas')
plt.xlabel('Fecha')
plt.xticks(rotation=45)
plt.legend(title='Categor√≠a')
plt.tight_layout()
plt.savefig("grafico_barras.png")  # Guardar imagen
plt.close()

# 4Ô∏è‚É£ üìÇ Guardar datos y gr√°ficos en un archivo Excel

# Crear un nuevo archivo Excel
wb = Workbook()
ws = wb.active
ws.title = "Resumen de Ventas"

# Guardar la tabla pivote en el archivo Excel
for i, col in enumerate(pivot_df.columns, start=2):
    ws.cell(row=1, column=i, value=col)

for i, (idx, row) in enumerate(pivot_df.iterrows(), start=2):
    ws.cell(row=i, column=1, value=idx.strftime('%Y-%m-%d'))
    for j, value in enumerate(row, start=2):
        ws.cell(row=i, column=j, value=value)

# Insertar im√°genes en Excel
img1 = Image("grafico_lineas.png")
img2 = Image("grafico_barras.png")
ws.add_image(img1, "E5")
ws.add_image(img2, "E20")

# Guardar el archivo
wb.save("reporte_ventas.xlsx")

print("‚úÖ Reporte generado: reporte_ventas.xlsx")
```
